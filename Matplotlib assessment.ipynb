# What is a scatter plot?
## Scatter plot is use to plot a   x and y coordinate in a graph.It uses dots to represent values for two different numeric variables

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("It is a X-axis",fontsize=16)
plt.ylabel("It is a Y-axis",fontsize=16)
plt.title("Scatter Plot",fontsize=14)
plt.scatter(x,y)
plt.show()
```

# Why is the subplot() function used? Draw four line plots using the subplot() function.
## The subplot() function takes three arguments that describes the layout of the figure.The layout is organized in rows and columns, which are represented by the first and second argument.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.figure(figsize=(15,8))
plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])



plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 3, 4)
plt.plot(x,y)


plt.show()
```

# What is a bar plot?
## A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel("Companies",fontsize=16)
plt.ylabel("Profit",fontsize=16)
plt.title("Net profit of companies",fontsize=14)

plt.bar(company,profit,color='g')
plt.show()
```

# What is a box plot? Why is it used?

### Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
import matplotlib.pyplot as plt
import random
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot(box1)
plt.boxplot(box2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
print()
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200) 
box2 = np.random.normal(90, 20, 200) 

data = [box1, box2]

fig, ax = plt.subplots(figsize=(8,6))
ax.boxplot(data)


ax.set_title('Box plot of box1 and box2')
ax.set_xlabel('Box')
ax.set_ylabel('Value')


plt.show()
```
